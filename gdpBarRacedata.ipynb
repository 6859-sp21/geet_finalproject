```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition,
                                                  mark_inset)
import pandas as pd
```

```python
data = pd.read_csv("GDP_billions2.csv")
```

```python
data.head()
data.drop('South Asia', inplace=True, axis=1)
data.head()
```

```python
# usonly = data.iloc[:,:2]
# usonly.head()
```

```python
rearranged = data.melt(id_vars=["date"], 
        var_name="Country", 
        value_name="Value")
```

```python
rearranged
```

```python
rearranged['lastValue'] = rearranged['Value'].shift(1)
```

```python
rearranged.loc[rearranged.date == 1960, "lastValue"] = 0
# rearranged[rearranged['Country'] == "China"]
# rearranged
```

```python
rearranged
```

```python
# rearranged[rearranged['Country'] == "India"]
```

```python
rearranged = rearranged.sort_values(["date", "Value"], ascending = [True, False])
rearranged
```

```python
##list of ranks from 1 to 185
ranks = [x+1 for x in range(185)]
rearranged['Rank']= pd.np.tile(ranks, len(rearranged) // len(ranks)).tolist() + ranks[:len(rearranged)%len(ranks)]
```
